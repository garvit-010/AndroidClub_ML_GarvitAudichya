```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
demo = pd.DataFrame(scaled_data,columns=df.columns)
```

```python
demo.describe()
```

```python
demo.info()
```

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
# Elbow points K=3 or K=4
```

```python

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_data)

df_selected['Cluster'] = kmeans.labels_

df_selected.head()
```

```python
cluster_summary = df_selected.groupby('Cluster').mean()
print(cluster_summary)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=df_selected['Cluster'], cmap='viridis', alpha=0.7)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Customer Segmentation Clusters')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(scaled_data, kmeans.labels_)
print(f"Silhouette Score for k=3: {score}")
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(scaled_data)

df_selected['Cluster'] = kmeans.labels_

df_selected.head()
```

```python
cluster_summary = df_selected.groupby('Cluster').mean()
print(cluster_summary)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=df_selected['Cluster'], cmap='viridis', alpha=0.7)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Customer Segmentation Clusters')
plt.colorbar(label='Cluster')
plt.show()
```
